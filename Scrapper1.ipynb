```python
import requests
from bs4 import BeautifulSoup
import urllib.parse
import os

HEADERS = {'User-Agent': 'Mozilla/5.0 (iPad; CPU OS 12_2 like Mac OS X) AppleWebKit/605.1.15 (KHTML, like Gecko) Mobile/15E148'}

def crawl(url, visited=None, max_depth=2, output_file="output.txt"):
    # if visited is None:
    #     visited = set()

    # if url in visited or max_depth == 0:
    #     return

    # visited.add(url)

    try:
        response = requests.get(url, headers=HEADERS)
        response.raise_for_status()
        soup = BeautifulSoup(response.content, "html.parser")

        with open(output_file, "a", encoding="utf-8") as f: #Open file in append mode.
            f.write(f"Crawling: {url}\n")
            paragraphs = soup.find_all("p")
            f.write("Paragraphs:\n")
            for paragraph in paragraphs:
                f.write(paragraph.get_text(strip=True) + "\n")

            f.write("\nHeadings:\n")
            headings = soup.find_all(["h1", "h2", "h3", "h4", "h5", "h6"])
            for heading in headings:
                f.write(heading.get_text(strip=True) + "\n")

            # links = soup.find_all("a")
            # for link in links:
            #     href = link.get("href")
            #     if href:
            #         absolute_url = urllib.parse.urljoin(url, href)
            #         crawl(absolute_url, visited, max_depth - 1, output_file) #Pass filename

    except requests.exceptions.RequestException as e:
        with open(output_file, "a", encoding="utf-8") as f:
            f.write(f"Error accessing {url}: {e}\n")
    except Exception as e:
        with open(output_file, "a", encoding="utf-8") as f:
            f.write(f"An unexpected error occurred: {e}\n")


# Start crawling and ensure the output file exists.
if not os.path.exists("output2.txt"):
    open("output2.txt", "w").close() # creates the file if it doesn't exist

base_url="https://segment.com"
visited=set()
for i in soup.find_all("a"):
  temp=i.get("href")
  if not temp:
    continue
  if "nav-list__link" in i.get("class", []) or "social-list__link" in i.get("class", []):
    continue
  if temp[0]=="/":
    temp=base_url+temp
  # print(temp)
  if temp not in visited:
    visited.add(temp)
    crawl(temp, max_depth=1,output_file="output2.txt")
    print("Done:",temp)
```
